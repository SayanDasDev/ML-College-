```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Loading CSV

```python
data = pd.read_csv('./data/data.csv')
data.head()
```

#### Removin Outlies and Making plot

```python
isOutlier = (data["Shoesize"]>10) | (data["Height"]<70)

data.drop(data[isOutlier].index ,axis=0 ,inplace=True)
data.head()
```

```python
fig, ax = plt.subplots(3, 1, figsize=(15,20))

data['Weight'].plot(kind='bar', ax=ax[0])
data['Shoesize'].plot(kind='bar', ax=ax[1])
data['Height'].plot(kind='bar', ax=ax[2])

ax[0].set_title('Weight')
ax[1].set_title('Shoesize')
ax[2].set_title('Height')
plt.show()
```
